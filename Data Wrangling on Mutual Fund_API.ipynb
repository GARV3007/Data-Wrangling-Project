```python
import urllib.request, urllib.parse
from urllib.error import HTTPError, URLError
import json
import pandas as pd
```

```python
def get_data():
    
    # Set URL
    url = 'https://financialmodelingprep.com/api/v3/quotes/mutual_fund?apikey=ad9104210bf1828a131927ab5e448c4a'
    
    try:
        uh = urllib.request.urlopen(url)        # Connect to API
    except HTTPError as e:                      # If HTTP error, display error and exit
        print("Sorry! Could not retrieve anything from API")
        return none
    except URLError as e:                       # If URL error, display erorr and exit
        print('Failed to reach a server.')
        print('Reason: ', e.reason)
        return None
    else:
        API_data = uh.read().decode()           # Good connection, read from API
        print("Successfully retrieved data from API")
        return API_data
```

```python
def build_dataframe(json_data):
    """
    Call to this function will build dataframe from json_data
    """
    
    # Define empty dictionary
    MF_dict = {'Symbol':[], 'name_a':[], 'price':[], 'changesPercentage':[], 'change_a':[], 'dayLow':[], 'dayHigh':[], 
              'yearHigh':[], 'yearLow':[], 'marketCap':[], 'priceAvg50':[], 'priceAvg200':[], 'volume':[], 'avgVolume':[],
              'exhange':[], 'open':[], 'previousClose':[], 'eps':[], 'pe':[], 'earningsAnnouncement':[], 
              'sharesOutstanding':[], 'timestamp':[]}
    
    # Loop for each dictionary element in JSON list
    for d in json_data:
        MF_dict['Symbol'].append(d['symbol'])
        MF_dict['name_a'].append(d['name'])
        MF_dict['price'].append(d['price'])
        MF_dict['changesPercentage'].append(d['changesPercentage'])
        MF_dict['change_a'].append(d['change'])
        MF_dict['dayLow'].append(d['dayLow'])
        MF_dict['dayHigh'].append(d['dayHigh'])
        MF_dict['yearHigh'].append(d['yearHigh'])
        MF_dict['yearLow'].append(d['yearLow'])
        MF_dict['marketCap'].append(d['marketCap'])
        MF_dict['priceAvg50'].append(d['priceAvg50'])
        MF_dict['priceAvg200'].append(d['priceAvg200'])
        MF_dict['volume'].append(d['volume'])
        MF_dict['avgVolume'].append(d['avgVolume'])
        MF_dict['sharesOutstanding'].append(d['sharesOutstanding'])
        MF_dict['timestamp'].append(d['timestamp'])
        

    df = pd.DataFrame.from_dict(MF_dict, orient='index')  # Build Dataframe from dictionary with index
    return df
        
        
```

```python
# Get Data from API
data = get_data()
```

```python
# Load data from API into JSON format
json_data = json.loads(data)
```

```python
# Build dataframe from JSON data
df = build_dataframe(json_data)
df.head()
```
